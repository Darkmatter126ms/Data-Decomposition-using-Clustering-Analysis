# Preprocessing of dataset

```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("bodyfat.csv")
df
```

```python
# Understanding the dataset
df.info()
```

```python
# Checking for empty rows
empty_rows = df.isnull().all(axis=1)
print(f"Empty rows: {empty_rows}")
```

```python
# Drop BodyFat column
X = df.drop(columns=["BodyFat"])
X
```

```python
# Standardizing input features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)
```

# Agglomerative Hierarchical Clustering (AHC)

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from mpl_toolkits.mplot3d import Axes3D
```

## Scatter plots before clustering

```python
pca2 = PCA(n_components=2)
X_pca2 = pca2.fit_transform(X_scaled)

pca3 = PCA(n_components=3)
X_pca3 = pca3.fit_transform(X_scaled)

# 2D and 3D scatter plot before clustering
fig = plt.figure(figsize=(14, 6))

ax2 = fig.add_subplot(1, 2, 1)
ax2.scatter(
    X_pca2[:,0],
    X_pca2[:,1],
    alpha=0.5
)
ax2.set_xlabel("PCA1")
ax2.set_ylabel("PCA2")
ax2.set_title("Body Fat - 2D PCA Scatter plot (before AHC)")

ax3 = fig.add_subplot(1, 2, 2, projection="3d")
ax3.scatter(
    X_pca3[:,0],
    X_pca3[:,1],
    X_pca3[:,2],
    alpha=0.5
)
ax3.set_xlabel("PCA1")
ax3.set_ylabel("PCA2")
ax3.set_zlabel("PCA3")
ax3.set_title("Body Fat - 3D Scatter plot (before AHC)")
plt.show()
```

## Training

### Determining the optimal number of clusters  
- Silhouette Scores  
- David-Bouldin Index (DBI)  
- Calinski-Harabasz Index (CHI))

```python
candidate_clusters = range(2, 50)
sil_score_list = []
dbi_score_list = []
chi_score_list = []

for n_clusters in candidate_clusters:
    ahc_clusters = AgglomerativeClustering(n_clusters=n_clusters, linkage="ward")
    labels_clusters = ahc_clusters.fit_predict(X_scaled)

    sil_score = silhouette_score(X_scaled, labels_clusters)
    dbi_score = davies_bouldin_score(X_scaled, labels_clusters)
    chi_score = calinski_harabasz_score(X_scaled, labels_clusters)

    sil_score_list.append(sil_score)
    dbi_score_list.append(dbi_score)
    chi_score_list.append(chi_score)

    print(f"Number of clusters: {n_clusters}, Silhouette score: {sil_score:.3f}, Davies-Bouldin score: {dbi_score:.3f}, Calinski-Harabasz score: {chi_score:.3f}")

max_sil = max(sil_score_list)
min_dbi = min(dbi_score_list)
max_chi = max(chi_score_list)

optimal_clusters_sil = candidate_clusters[sil_score_list.index(max_sil)]
optimal_clusters_dbi = candidate_clusters[dbi_score_list.index(min_dbi)]
optimal_clusters_chi = candidate_clusters[chi_score_list.index(max_chi)]

print()
print(f"Highest silhouette score: {max_sil:.3f}, Optimal number of clusters: {optimal_clusters_sil}")
print(f"Lowest Davies-Bouldin score: {min_dbi:.3f}, Optimal number of clusters: {optimal_clusters_dbi}")
print(f"Highest Calinski-Harabasz: {max_chi:.3f}, Optimal number of clusters: {optimal_clusters_chi}")
```

```python
# Plot a graph of silhouette scores against number of clusters
plt.figure(figsize=(10, 8))
plt.plot(candidate_clusters, sil_score_list, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette score")
plt.title("Body Fat - Silhouette scores against Number of clusters")
plt.show()
```

```python
# DBI scores against number of clusters
plt.figure(figsize=(10, 8))
plt.plot(candidate_clusters, dbi_score_list, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Davies-Bouldin score")
plt.title("Body Fat - Davies-Bouldin scores against Number of clusters")
plt.show()
```

```python
# CHI scores against number of clusters
plt.figure(figsize=(10, 8))
plt.plot(candidate_clusters, chi_score_list, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Calinski-Harabasz score")
plt.title("Body Fat - Calinski-Harabasz scores against Number of clusters")
plt.show()
```

## Experimental analysis

### Analysis of AHC model performance on different linkage criterias
(using optimal number of clusters)    
- ward: minimizes the variance of the clusters being merged 
- single: uses the min distances between all observations of the 2 sets  
- complete: uses max distances between all observations of the 2 sets  
- average: uses the average of the distances of each observation of the 2 sets

References:  
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html 

```python
n_clusters_optimal = 2 # Change based on the results given 4 code blocks above
linkage_methods = ["ward","single", "complete", "average"]
sil_linkage_list = []
dbi_linkage_list = []
chi_linkage_list = []

for m in linkage_methods:
    
    print(f"Linkage method: {m.capitalize()}")

    # Conduct AHC
    ahc_linkage = AgglomerativeClustering(n_clusters=n_clusters_optimal, linkage=m)
    labels_linkage = ahc_linkage.fit_predict(X_scaled)

    # Plot 2D PCA scatter plot
    fig_linkage = plt.figure(figsize=(14, 6))
    ax2_linkage = fig_linkage.add_subplot(1, 2, 1)
    
    for c in np.unique(labels_linkage):
        ax2_linkage.scatter(
            X_pca2[labels_linkage == c, 0], 
            X_pca2[labels_linkage == c, 1], 
            label=f"Cluster {c}", 
            alpha=0.5
        )

    ax2_linkage.set_xlabel("PCA1")
    ax2_linkage.set_ylabel("PCA2")
    ax2_linkage.set_title(f"Body Fat - 2D Scatter plot (after AHC + {m.capitalize()} linkage)")
    ax2_linkage.legend()

    # Plot 3D PCA scatter plot
    ax3_linkage = fig_linkage.add_subplot(1, 2, 2, projection="3d")
    
    for c in np.unique(labels_linkage):
        ax3_linkage.scatter(
            X_pca3[labels_linkage == c, 0], 
            X_pca3[labels_linkage == c, 1], 
            X_pca3[labels_linkage == c, 2], 
            label=f"Cluster {c}", 
            alpha=0.5
        )

    ax3_linkage.set_xlabel("PCA1")
    ax3_linkage.set_ylabel("PCA2")
    ax3_linkage.set_zlabel("PCA3")
    ax3_linkage.set_title(f"Body Fat - 3D Scatter plot (after AHC + {m.capitalize()} linkage)")
    # plt.show()

    # Plot dendrogram for each linkage method
    Z_linkage = linkage(X_scaled, method=m)

    plt.figure(figsize=(10, 8))
    dendrogram(Z_linkage)
    plt.xlabel("Data points")
    plt.ylabel("Distance")
    plt.title(f"Body Fat - Dendrogram (after AHC + {m.capitalize()} linkage)")
    plt.show()
    print("-------------------------------------------------------------------------")

    # Silhouette score, DBI score, CHI score 
    sil_linkage = silhouette_score(X_scaled, labels_linkage)
    dbi_linkage = davies_bouldin_score(X_scaled, labels_linkage)
    chi_linkage = calinski_harabasz_score(X_scaled, labels_linkage)

    sil_linkage_list.append(sil_linkage)
    dbi_linkage_list.append(dbi_linkage)
    chi_linkage_list.append(chi_linkage)

print()
print("Linkage method | Silhouette score | Davis-Bouldin score | Calinski-Harabasz score")
for i, m in enumerate(linkage_methods):
    print(f"{m:<12} | {sil_linkage_list[i]:.3f} | {dbi_linkage_list[i]:.3f} | {chi_linkage_list[i]:.3f}")
# Should still rely on dendrogram and PCA scatter plots to determine best linkage method as dendrogram visualizes the full hierarchical process
```

### Analysis of AHC model performance on different number of principle components 
(using optimal number of clusters and ward linkage)

```python
n_features = X_scaled.shape[1] # Number of feature columns in dataset
pca_comps = list(range(2, n_features+1))

sil_pca_list = []
dbi_pca_list = []
chi_pca_list = []

for n in pca_comps:
    pca = PCA(n_components=n)
    X_pca = pca.fit_transform(X_scaled)

    # AHC 
    ahc_pca = AgglomerativeClustering(n_clusters=n_clusters_optimal, linkage="ward")
    labels_pca = ahc_pca.fit_predict(X_pca)

    # Scores
    sil_pca = silhouette_score(X_pca, labels_pca)
    dbi_pca = davies_bouldin_score(X_pca, labels_pca)
    chi_pca = calinski_harabasz_score(X_pca, labels_pca)

    sil_pca_list.append(sil_pca)
    dbi_pca_list.append(dbi_pca)
    chi_pca_list.append(chi_pca)

# Results
print(f"Number of components | Silhouette score | Davies-Bouldin score | Calinski-Harabasz score")
for i, n in enumerate(pca_comps):
    print(f"{n:<21} | {sil_pca_list[i]:.3f} | {dbi_pca_list[i]:.3f} | {chi_pca_list[i]:.3f}")

best_sil_pca = max(sil_pca_list)
best_dbi_pca = min(dbi_pca_list)
best_chi_pca = max(chi_pca_list)

n_pc_sil = pca_comps[sil_pca_list.index(best_sil_pca)]
n_pc_dbi = pca_comps[dbi_pca_list.index(best_dbi_pca)]
n_pc_chi = pca_comps[chi_pca_list.index(best_chi_pca)]

print()
print(f"Highest Silhouette score: {best_sil_pca:.3f}, Optimal number of components: {n_pc_sil}")
print(f"Lowest Davies-Bouldin score: {best_dbi_pca:.3f}, Optimal number of components: {n_pc_dbi}")
print(f"Highest Calinski-Harabasz score: {best_chi_pca:.3f}, Optimal number of components: {n_pc_chi}")
```

```python
# Plot results
plt.figure(figsize=(10, 6))
plt.plot(pca_comps, sil_pca_list, marker="o")
plt.xlabel("Number of PCA components")
plt.ylabel("Silhouette score")
plt.title("Body Fat - AHC performance against Number of PCA components (Silhouette)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(pca_comps, dbi_pca_list, marker="o")
plt.xlabel("Number of PCA components")
plt.ylabel("Davies-Bouldin score")
plt.title("Body Fat - AHC performance against Number of PCA components (Davies-Bouldin)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(pca_comps, chi_pca_list, marker="o")
plt.xlabel("Number of PCA components")
plt.ylabel("Calinski-Harabasz score")
plt.title("Body Fat - AHC performance against Number of PCA components (Calinski-Harabasz)")
plt.show()
```

```python
# Plotting variance against number of PCA components
pca_full = PCA().fit(X_scaled)
explained_var = np.cumsum(pca_full.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_var)+1), explained_var, marker="o")
plt.xlabel("Number of PCA components")
plt.ylabel("Cumulative explained variance")
plt.title("Body Fat - Explained variance against Number of PCA components")
plt.show()
```

## Final AHC on the optimal configuration
- number of clusters 
- number of principle components 
- linkage criteria

```python
pca_final = PCA(n_components=2) # change according to the results from the previous code cell
X_pca_final = pca_final.fit_transform(X_scaled)

ahc_final = AgglomerativeClustering(n_clusters=n_clusters_optimal, linkage="ward")
labels_final = ahc_final.fit_predict(X_pca_final)
```

```python
# Plot final 2D scatter plot
plt.figure(figsize=(8, 6))

plt.scatter(
        X_pca_final[:,0],
        X_pca_final[:,1], 
        c=labels_final, 
        cmap="tab10", 
        s=50
    )

plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("Body Fat - Final AHC (Ward linkage + PCA reduced)")
plt.colorbar(label="Cluster")

plt.show()
```

```python
# Plot dendrogram
Z_linkage = linkage(X_pca_final, method="ward")

plt.figure(figsize=(10, 8))
dendrogram(Z_linkage)
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.title(f"Body Fat - Dendrogram (Ward linkage + PCA reduced)")
plt.show()
```

```python
# Cluster analysis
df["Cluster"] = labels_final # Add cluster labels to df
cluster_summary = df.groupby("Cluster").mean().round(2)
print(cluster_summary)
```

```python
# Heatmap for cluster analysis
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(cluster_summary, annot=True, cmap="YlGnBu", fmt=".2f", cbar=True)
plt.title("Body Fat - Heatmap of average feature values by cluster", fontsize=14, pad=12)
plt.xlabel("Feature")
plt.ylabel("Cluster")
plt.tight_layout()
plt.show()
```
